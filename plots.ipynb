```python
import json
with open('./results/2021-09-24-15-01-50.json', mode='r') as fp:
    results = json.load(fp)
remarks = [x['remarks'] for x in results]
pings = [x['ping'] for x in results]
gpings = [x['gPing'] for x in results]
dspeeds = [x['dspeed'] for x in results]
maxspeeds = [x['maxDSpeed'] for x in results]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({'ping': pings, 'gping': gpings, 'remark': remarks})
plt.figure(figsize=(20,10))
sns.barplot(x = 'remark', y = 'gping', data = df, color = 'blue', label='Google Ping')
sns.barplot(x = 'remark', y = 'ping', data = df, color = 'red', alpha=0.8, label='Ping')
```
